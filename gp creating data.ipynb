```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
import os 
```

```python
DataDir = r"F:\\final project\\Data_set" 
Categorize = ["Green_1","Red_1","Yellow_1","None_1"]

for category in Categorize: 
    path = os.path.join(DataDir,category)
    for img in os.listdir(path):
        image_array = cv2.imread(os.path.join(path,img))
        plt.imshow(image_array)
        plt.show()
        break
    break
```

```python
print(image_array.shape)
```

```python
image_hight = 80 
image_weidth = 60

new_array = cv2.resize(image_array, (image_hight,image_weidth)) 
plt.imshow(new_array)
plt.show()
```

```python
training_Data = []
def create_data_set():
    for category in Categorize:
        path = os.path.join(DataDir,category)
        class_num = Categorize.index(category)
        for img in os.listdir(path):
            try: 
                image_array = cv2.imread(os.path.join(path,img), 1)
                new_array = cv2.resize(image_array,(image_hight,image_weidth))
                training_Data.append([new_array, class_num])
            except Exception as e:
                pass

            
create_data_set() 
```

```python
print(len(training_Data)) 
```

```python
import random 

random.shuffle(training_Data) 
```

```python
for sample in training_Data[:10]: 
    print(sample[1])
```

```python
X = []  
y = [] 
```

```python
for feature, label in training_Data:
    X.append(feature) 
    y.append(label)
```

```python
X = np.array(X).reshape(-1,image_hight,image_weidth,3) 
y = np.array(y)
```

```python
import pickle 

pickle_out  = open(r"F:\\final project\\Data_set\\X.pickle", "wb")
pickle.dump(X,pickle_out) 
pickle_out.close()

pickle_out  = open(r"F:\\final project\\Data_set\\y.pickle", "wb")
pickle.dump(y,pickle_out)
pickle_out.close()
```
