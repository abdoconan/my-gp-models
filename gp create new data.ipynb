```python
import cv2 
import numpy as np 
import os 
import matplotlib.pyplot as plt
```

```python
DataDir = r"F:\\final project\\Data_set" #directory of data
Categorize = ["Green_1","Red_1","Yellow_1","None_1"]

for category in Categorize: #read data from path directory 
    path = os.path.join(DataDir,category)
    w = len(os.listdir(path)) -1
    for img in os.listdir(path):
        image_array = cv2.imread(os.path.join(path,img))
        flip = cv2.flip(image_array,1)
        cv2.imwrite(get_path(w,category),flip)
        w +=1
    
```

```python
def get_path(j , category):
    x = os.path.join(DataDir,category)
    imageSave=  os.path.join(x , str(j))
    extention = ".jpg"
    path = imageSave + extention
    print(path)
    return path
```
