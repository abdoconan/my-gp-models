```python
pickle_in = open('generator_model.pickle','rb')
model = pickle.load(pickle_in)
pickle_in.close()
```

```python
image = cv2.imread(path+"/1_e0UlsRVfTM2xw_uVWTsPVg.png")
haar = cv2.CascadeClassifier(path+"/haarCascade.xml")
```

```python
def grayscale(img):
    return cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

def equalize(img):
    return cv2.equalizeHist(img)


def preprocessing(img):
    img = grayscale(img)
    img = equalize(img)
    img = img/255
    return img
```

```python
def detect_sign(image):
    gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
    cv2.imshow("70" , image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.imshow("70" , gray)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    sign = haar.detectMultiScale(gray,1.2,5)
    if len(sign)>0:
        for (x,y,w,h) in sign:    
            x = x-50
            y = y-50
            w = w +50
            h = h + 50
            cv2.rectangle(image , (x,y),(x+w,y+h),(0,0,0),2)
            img = image[y:y+h,x:x+w]
        cv2.imshow("90" , image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        cv2.imshow("70" , img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        img = np.asarray(img)
        img = cv2.resize(img,(32,32))
        img = preprocessing(img)
        img = img.reshape(1,32,32,1)
        return img
```
