```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D ,MaxPooling2D ,Dropout ,Dense ,Activation ,Flatten
import pickle
```

```python
pickle_in = open(r"F:\\final project\\Data_set\\X.pickle","rb") 
X = pickle.load(pickle_in)                                       
pickle_in.close()

pickle_in = open(r"F:\\final project\\Data_set\\y.pickle","rb")
y = pickle.load(pickle_in)
pickle_in.close()
```

```python
tf.keras.utils.normalize(X) 
```

```python
print(X[1].shape) 
```

```python
model = Sequential() 
model.add(Conv2D(32,(3,3), input_shape = X.shape[1:]))  
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.2))



model.add(Conv2D(64,(3,3))) 
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.2))


model.add(Conv2D(132,(3,3))) 
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.2))

model.add(Flatten())
model.add(Dense(132)) 
model.add(Activation("relu"))



model.add(Dense(132))
model.add(Activation("relu"))

model.add(Dense(4)) 
model.add(Activation("sigmoid")) 

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam' ,
              metrics = ['accuracy']) 
```

```python
model.fit(X,y, batch_size= 32, epochs =2 , validation_split=0.1) #fit data #validation dy bt2sm el data le 90% training data 
                                                                 #10% validation set (test set)
```
